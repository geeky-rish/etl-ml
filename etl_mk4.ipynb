```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import urllib.request
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans
from sklearn.ensemble import IsolationForest
from pymongo import MongoClient
from pymongo.errors import ConnectionFailure
import warnings

warnings.filterwarnings("ignore")

# MongoDB Configuration
MONGO_URI = "mongodb+srv://<etl_mk4>:<etl_mark4>@first.t3tgs.mongodb.net/?retryWrites=true&w=majority&appName=first"
DATABASE_NAME = "ETL_Project"
COLLECTION_NAME = "ProcessedData"

# Connect to MongoDB
def connect_mongo():
    try:
        client = MongoClient(MONGO_URI)
        db = client[DATABASE_NAME]
        print("‚úÖ Connected to MongoDB successfully!")
        return db
    except ConnectionFailure as e:
        print(f"‚ùå MongoDB Connection Failed: {e}")
        return None

# Fetch Dataset from URL
def fetch_data():
    dataset_url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
    local_file = "data.csv"

    try:
        urllib.request.urlretrieve(dataset_url, local_file)
        print("‚úÖ Dataset fetched successfully!")
        df = pd.read_csv(local_file)
        return df
    except Exception as e:
        print(f"‚ùå Data Fetching Failed: {e}")
        return None

# Data Cleaning
def clean_data(df):
    try:
        # Drop unwanted columns
        df.drop(["Name", "Ticket", "Cabin"], axis=1, inplace=True)

        # Convert categorical columns
        label_cols = ["Sex", "Embarked"]
        for col in label_cols:
            df[col] = LabelEncoder().fit_transform(df[col].astype(str))

        # Handle missing values using KNN Imputer
        imputer = KNNImputer(n_neighbors=5)
        df.iloc[:, :] = imputer.fit_transform(df)

        print("‚úÖ Data Cleaning Successful!")
        return df
    except Exception as e:
        print(f"‚ùå Data Cleaning Failed: {e}")
        return None

# Data Transformation
def transform_data(df):
    try:
        # Standardize the data
        scaler = StandardScaler()
        df.iloc[:, :] = scaler.fit_transform(df)

        print("‚úÖ Data Transformation Successful!")
        return df
    except Exception as e:
        print(f"‚ùå Data Transformation Failed: {e}")
        return None

# Optuna Objective Function
def objective(trial):
    try:
        # Suggest values for hyperparameters
        n_clusters = trial.suggest_int("n_clusters", 2, 10)  # Tune number of clusters
        contamination = trial.suggest_float("contamination", 0.01, 0.2)  # Tune outlier percentage

        # Load and preprocess data
        df = fetch_data()
        df = clean_data(df)
        df = transform_data(df)

        # Outlier Removal
        iso = IsolationForest(contamination=contamination, random_state=42)
        df["Anomaly"] = iso.fit_predict(df)
        df = df[df["Anomaly"] == 1]  # Keep only normal data
        df.drop(columns=["Anomaly"], inplace=True)

        # Clustering
        kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
        df['Cluster'] = kmeans.fit_predict(df)

        return -kmeans.inertia_  # Minimize inertia for better clustering

    except Exception as e:
        print(f"‚ùå Optuna Optimization Failed: {e}")
        return float('inf')  # Return a large value to avoid stopping optimization

# Run Optuna optimization
study = optuna.create_study(direction="minimize")  # Minimize inertia
study.optimize(objective, n_trials=20)  # Run 20 trials

# Get best hyperparameters
best_params = study.best_params
n_clusters = best_params["n_clusters"]
contamination = best_params["contamination"]

print(f"‚úÖ Best Parameters Found: {best_params}")

# Outlier Detection using Optimized Isolation Forest
def detect_outliers(df):
    try:
        iso = IsolationForest(contamination=contamination, random_state=42)
        df["Anomaly"] = iso.fit_predict(df)
        df = df[df["Anomaly"] == 1]  # Keep normal data
        df.drop(columns=["Anomaly"], inplace=True)

        print(f"‚úÖ Outliers removed successfully using contamination={contamination}!")
        return df
    except Exception as e:
        print(f"‚ùå Outlier Detection Failed: {e}")
        return None

# Clustering using Optimized K-Means
def cluster_data(df):
    try:
        kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
        df['Cluster'] = kmeans.fit_predict(df)

        print(f"‚úÖ Clustering Successful with n_clusters={n_clusters}!")
        return df
    except Exception as e:
        print(f"‚ùå Clustering Failed: {e}")
        return None


# Save to MongoDB
def save_to_mongo(db, df):
    try:
        records = df.to_dict(orient="records")
        db[COLLECTION_NAME].insert_many(records)
        print("‚úÖ Data saved to MongoDB successfully!")
    except Exception as e:
        print(f"‚ùå Data Saving Failed: {e}")

# Data Visualization
def visualize_data(df):
    try:
        sns.set_style("whitegrid")

        # 1. Cluster Distribution
        plt.figure(figsize=(8, 5))
        sns.countplot(x="Cluster", data=df, palette="coolwarm")
        plt.title("Cluster Distribution")
        plt.xlabel("Cluster")
        plt.ylabel("Count")
        plt.show()

        # 2. Correlation Heatmap
        plt.figure(figsize=(10, 6))
        sns.heatmap(df.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
        plt.title("Feature Correlation Heatmap")
        plt.show()

        # 3. Histogram of Features
        df.hist(figsize=(12, 8), bins=30, color="skyblue")
        plt.suptitle("Feature Distributions")
        plt.show()

        # 4. Pairplot of Clusters
        sns.pairplot(df, hue="Cluster", palette="coolwarm")
        plt.show()

        print("‚úÖ Visualization Complete!")
    except Exception as e:
        print(f"‚ùå Visualization Failed: {e}")

# Statistical Analysis
def statistical_analysis(df):
    try:
        print("\nüìä **Statistical Analysis:**")
        print(df.describe())

        print("\nüî¢ **Feature Correlation:**")
        print(df.corr())

        print("‚úÖ Statistical Analysis Complete!")
    except Exception as e:
        print(f"‚ùå Statistical Analysis Failed: {e}")

# Main Function
def main():
    db = connect_mongo()
    if db is None:
        return

    df = fetch_data()
    if df is None:
        return

    df = clean_data(df)
    if df is None:
        return

    df = transform_data(df)
    if df is None:
        return

    df = detect_outliers(df)
    if df is None:
        return

    df = cluster_data(df)
    if df is None:
        return

    save_to_mongo(db, df)
    visualize_data(df)
    statistical_analysis(df)

if __name__ == "__main__":
    main()
```

```python
# -------------------- IMPORTS -------------------- #
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import joblib
import logging
import optuna
from io import BytesIO, StringIO
import base64
from datetime import datetime
from typing import Tuple, Dict, Any
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.cluster import KMeans
from sklearn.ensemble import IsolationForest
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from flask import Flask, render_template_string, send_file
from pyngrok import ngrok
import warnings
warnings.filterwarnings("ignore")

# -------------------- CONFIGURATION -------------------- #
class Config:
    DATA_URL = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
    NUMERICAL_FEATURES = ['Age', 'Fare', 'SibSp', 'Parch']
    CATEGORICAL_FEATURES = ['Sex', 'Embarked', 'Pclass']
    DROP_FEATURES = ['Name', 'Ticket', 'Cabin', 'PassengerId']
    OPTUNA_TRIALS = 10
    CONTAMINATION_RANGE = (0.05, 0.2)
    N_CLUSTERS_RANGE = (2, 6)
    PORT = 5001
    ARTIFACT_PATH = "/content/artifacts"
    PLOT_STYLE = "whitegrid"
    COLOR_PALETTE = "viridis"

# -------------------- ENHANCED PIPELINE -------------------- #
class EnhancedETL:
    def __init__(self):
        self.preprocessor = None
        self.best_params = None

    def _create_preprocessor(self) -> ColumnTransformer:
        numerical_pipe = Pipeline([
            ('imputer', KNNImputer(n_neighbors=5)),
            ('scaler', StandardScaler())
        ])

        categorical_pipe = Pipeline([
            ('encoder', OneHotEncoder(handle_unknown='ignore'))
        ])

        return ColumnTransformer([
            ('num', numerical_pipe, Config.NUMERICAL_FEATURES),
            ('cat', categorical_pipe, Config.CATEGORICAL_FEATURES)
        ])

    def optimize_hyperparameters(self, X: pd.DataFrame) -> Dict[str, Any]:
        def objective(trial: optuna.Trial) -> float:
            contamination = trial.suggest_float('contamination', *Config.CONTAMINATION_RANGE)
            n_clusters = trial.suggest_int('n_clusters', *Config.N_CLUSTERS_RANGE)

            X_processed = self.preprocessor.fit_transform(X)
            iso = IsolationForest(contamination=contamination)
            mask = iso.fit_predict(X_processed) == 1
            kmeans = KMeans(n_clusters=n_clusters, n_init=10)
            kmeans.fit(X_processed[mask])
            return kmeans.inertia_

        study = optuna.create_study(direction='minimize')
        study.optimize(objective, n_trials=Config.OPTUNA_TRIALS)
        self.best_params = study.best_params
        return self.best_params

    def run_pipeline(self, df: pd.DataFrame) -> pd.DataFrame:
        self.preprocessor = self._create_preprocessor()
        best_params = self.optimize_hyperparameters(df)
        X_processed = self.preprocessor.fit_transform(df)

        iso = IsolationForest(contamination=best_params['contamination'])
        mask = iso.fit_predict(X_processed) == 1
        df_clean = df[mask].copy()

        kmeans = KMeans(n_clusters=best_params['n_clusters'], n_init=10)
        df_clean['Cluster'] = kmeans.fit_predict(self.preprocessor.transform(df_clean))

        joblib.dump(self.preprocessor, f"{Config.ARTIFACT_PATH}/preprocessor.joblib")
        df_clean.to_parquet(f"{Config.ARTIFACT_PATH}/processed_data.parquet")
        df_clean.to_csv(f"{Config.ARTIFACT_PATH}/processed_data.csv", index=False)

        return df_clean

# -------------------- PROFESSIONAL DASHBOARD -------------------- #
app = Flask(__name__)
sns.set_style(Config.PLOT_STYLE)
sns.set_palette(Config.COLOR_PALETTE)
processed_data = None

def create_plot(figsize=(10,6)):
    plt.figure(figsize=figsize)
    plt.tight_layout()
    return plt

@app.route('/')
def dashboard():
    return render_template_string('''
    <!DOCTYPE html>
    <html>
    <head>
        <title>AI ETL Dashboard</title>
        <link href="https://cdn.jsdelivr.net/npm/bootstrap@5.1.3/dist/css/bootstrap.min.css" rel="stylesheet">
        <style>
            .card { margin: 20px 0; box-shadow: 0 4px 8px rgba(0,0,0,0.1); }
            .plot-img { width: 100%; height: auto; border-radius: 8px; }
            .stats-card { background: #f8f9fa; padding: 20px; border-radius: 8px; }
        </style>
    </head>
    <body class="container mt-4">
        <h1 class="mb-4 text-center">AI-Driven ETL Analysis Dashboard</h1>

        <div class="row">
            <div class="col-md-12">
                <div class="card">
                    <div class="card-body">
                        <h4 class="card-title">Quick Actions</h4>
                        <a href="/cluster_analysis" class="btn btn-primary">Cluster Analysis</a>
                        <a href="/distribution" class="btn btn-info">Distributions</a>
                        <a href="/correlation" class="btn btn-success">Correlations</a>
                        <a href="/download" class="btn btn-secondary">Download Data</a>
                    </div>
                </div>
            </div>
        </div>

        <div class="row">
            <div class="col-md-12">
                <div class="stats-card">
                    <h4>Key Statistics</h4>
                    {{ stats_table|safe }}
                </div>
            </div>
        </div>
    </body>
    </html>
    ''', stats_table=processed_data.describe().to_html(classes="table table-striped"))

@app.route('/cluster_analysis')
def cluster_analysis():
    try:
        plt = create_plot((12,6))
        g = sns.pairplot(processed_data, hue='Cluster', corner=True)
        g.fig.suptitle("Cluster Pairwise Relationships", y=1.02)
        img = BytesIO()
        plt.savefig(img, format='png', bbox_inches='tight')
        img.seek(0)
        return render_template(f'<img class="plot-img" src="data:image/png;base64,{base64.b64encode(img.getvalue()).decode()}">')
    except Exception as e:
        return error_handler(e)

@app.route('/distribution')
def distribution_analysis():
    try:
        plt = create_plot()
        fig, axes = plt.subplots(2, 2, figsize=(12,8))

        # Age Distribution
        sns.histplot(processed_data['Age'], kde=True, ax=axes[0,0])
        axes[0,0].set_title('Age Distribution')

        # Fare Distribution
        sns.boxplot(x=processed_data['Fare'], ax=axes[0,1])
        axes[0,1].set_title('Fare Distribution')

        # Cluster Distribution
        sns.countplot(x='Cluster', data=processed_data, ax=axes[1,0])
        axes[1,0].set_title('Cluster Distribution')

        # Class Distribution
        sns.countplot(x='Pclass', data=processed_data, ax=axes[1,1])
        axes[1,1].set_title('Class Distribution')

        img = BytesIO()
        plt.savefig(img, format='png', bbox_inches='tight')
        img.seek(0)
        return render_image(img)
    except Exception as e:
        return error_handler(e)

@app.route('/correlation')
def correlation_analysis():
    try:
        plt = create_plot((10,8))
        corr_matrix = processed_data.corr()
        sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
        plt.title("Feature Correlation Matrix")
        return render_image(plt)
    except Exception as e:
        return error_handler(e)

@app.route('/download')
def download():
    try:
        return send_file(
            f"{Config.ARTIFACT_PATH}/processed_data.csv",
            mimetype='text/csv',
            as_attachment=True,
            download_name=f"processed_data_{datetime.now().strftime('%Y%m%d')}.csv"
        )
    except Exception as e:
        return error_handler(e)

def render_image(plt):
    img = BytesIO()
    plt.savefig(img, format='png', bbox_inches='tight')
    img.seek(0)
    plt.close()
    return f'<img class="plot-img" src="data:image/png;base64,{base64.b64encode(img.getvalue()).decode()}">'

def error_handler(e):
    return f'''
    <div class="alert alert-danger">
        <h4>Error Occurred</h4>
        <pre>{str(e)}</pre>
        <a href="/" class="btn btn-primary">Return to Dashboard</a>
    </div>
    '''

# -------------------- MAIN EXECUTION -------------------- #
def main():
    global processed_data

    # Colab setup
    !mkdir -p {Config.ARTIFACT_PATH}
    !pip install -q flask pyngrok optuna scikit-learn pyarrow
    !ngrok authtoken 2twHHIhppxwLpGPDYSu6Gxs8zUE_4qhDG5cJsWk8H3J6wSut3

    # Run pipeline
    raw_data = pd.read_csv(Config.DATA_URL).drop(Config.DROP_FEATURES, axis=1)
    processed_data = EnhancedETL().run_pipeline(raw_data)

    # Start server
    public_url = ngrok.connect(Config.PORT).public_url
    print(f"üåê Dashboard URL: {public_url}")
    app.run(host='0.0.0.0', port=Config.PORT)

if __name__ == "__main__":
    main()
```
